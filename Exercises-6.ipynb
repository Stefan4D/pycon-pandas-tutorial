### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
col = sales2['Title'].str.extract(r'\((.*)\)').bfill()
sales2['Currency'] = col
ss2 = sales2.fillna('')
# ss2 = sales2.copy()
# ss2
```

```python
sales1_1 = sales1.rename(columns={ "Book title": "Title", "Number sold": "Units sold", "Sales price": "List price", "Royalty paid": "Royalty" })
sales1_1['Currency'] = 'USD'
```

```python

# ss2.fillna('')
rows_to_drop = ss2[ss2['Units sold'] == ''].index
rows_to_drop
```

```python
ss2.drop(rows_to_drop, inplace=True)
ss2
```

```python
ss2['Revenue'] = ss2['Units sold'] * ss2['List price']
sales1_1['Revenue'] = sales1_1['Units sold'] * sales1_1['List price']


sales3 = pd.concat([ss2, sales1_1], ignore_index=True)
sales3
```

```python
# sales3.groupby(['Title', 'Currency'])[['Units sold', 'List price', 'Royalty']].sum()
```

```python
sales3['Title'] = sales3['Title'].str.title()
sales3['Title'] = sales3['Title'].str.replace("'S", "'s")
sales3['Title'] = sales3['Title'].str.replace("’S", "'s")
sales4 = sales3.groupby(['Title', 'Currency'])[['Units sold', 'List price', 'Royalty', 'Revenue']].sum()
sales4
```
